```python
import random

# Step 1: Generate a list of workers with basic info (ID, name, gender, salary)
workers = []
for i in range(400):
    worker = {
        "id": i + 1,
        "name": f"Worker_{i+1}",
        "gender": random.choice(["male", "female"]),
        "salary": random.randint(5000, 30000)  # Random salary between $5,000 and $30,000
    }
    workers.append(worker)
```

```python
# Step 2: Generate payment slips and set employee levels based on conditions
for worker in workers:
    try:
        # Default employee level
        employee_level = "General"

        # Apply conditions for employee level
        if 10000 < worker["salary"] < 20000:
            employee_level = "A1"
        elif 7500 < worker["salary"] < 30000 and worker["gender"] == "female":
            employee_level = "A5-F"

        # Payment slip details
        print(f"Payment Slip for {worker['name']} (ID: {worker['id']})")
        print(f"  Gender: {worker['gender']}")
        print(f"  Salary: ${worker['salary']}")
        print(f"  Employee Level: {employee_level}")
        print("-" * 30)
    except Exception as e:
        print(f"Error processing worker {worker['id']}: {e}")
```

```python
# Step 3: Save to a file (optional)
with open("payment_slips.txt", "w") as file:
    for worker in workers:
        try:
            # Default employee level
            employee_level = "General"

            # Apply conditions for employee level
            if 10000 < worker["salary"] < 20000:
                employee_level = "A1"
            elif 7500 < worker["salary"] < 30000 and worker["gender"] == "female":
                employee_level = "A5-F"

            # Write payment slip to file
            file.write(f"Payment Slip for {worker['name']} (ID: {worker['id']})\n")
            file.write(f"  Gender: {worker['gender']}\n")
            file.write(f"  Salary: ${worker['salary']}\n")
            file.write(f"  Employee Level: {employee_level}\n")
            file.write("-" * 30 + "\n")
        except Exception as e:
            file.write(f"Error processing worker {worker['id']}: {e}\n")
```
